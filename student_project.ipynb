```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.histplot(df1['exam_score'], bins=30, kde=True)
plt.title('Distribution of Exam Scores')
plt.show()
```

Most students score between 60-80, which is a healthy middle range.
There's a noticeable spike at 100, a few students are scoring perfect, which is a bit unusual.
The distribution is slightly left skewed, more students score higher than the average rather than lower.
Very few students score below 30.




```python
sns.lineplot(data=df1, x='study_hours_per_day', y='exam_score')
plt.title('Study Hours vs Exam Score')
plt.show()
```

There is a strong positive relationship between study hours and exam scores. As study hours increase, exam scores rise consistently. Students studying 0-1 hours score mostly between 20-55, while those studying 6+ hours almost always score above 85, with many getting 100. This shows that study hours is one of the strongest predictors of performance in this dataset.

```python
sns.lineplot(data=df1, x='social_media_hours', y='exam_score')
plt.title('Social Media Hours vs Exam Score')
plt.show()
```


There is a mild negative relationship between social media hours and exam scores. Students using social media 0-1 hours tend to average around 75, while those using it 5+ hours drop to around 50-60. However, the wide shaded area (confidence interval) tells us there's a lot of variation, social media alone doesn't determine the score. Some students use social media a lot and still score well. It's a contributing factor but not as strong a predictor as study hours.

```python
sns.scatterplot(data=df1, x='attendance_percentage', y='exam_score')
plt.title('Attendance vs Exam Score')
plt.show()
```

From the above observation we can see that attendance doesn't have much of a relationship with the exam scores as there are students whose attendance is low but still they score a 100 and vice versa.

```python
numeric_cols = df1.select_dtypes(include='number').drop(columns='age')

sns.heatmap(numeric_cols.corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap')
plt.show()
```

Study hours dominates everything at 0.83. Mental health is a distant but interesting second at 0.32. Everything else is relatively weak. This tells you that your model will likely rely heavily on study hours.

```python
df1.groupby('part_time_job')['exam_score'].mean().plot(kind='bar')
plt.title('Avg Exam Score by Part Time Job')
plt.show()
```

Students with and without a part time job score almost the same, around 70. The difference is negligible. This tells us that having a part time job has virtually no impact on exam performance in this dataset.

```python
df1.groupby('diet_quality')['exam_score'].mean().plot(kind='bar')
plt.title('Avg Exam Score by Diet Quality')
plt.show()
```

Diet quality shows almost no difference in exam scores across Fair, Good, and Poor categories all three average around 69-71. Similar to part time job, diet quality appears to have very little impact on exam performance.

```python
df1.groupby('parental_education_level')['exam_score'].mean().plot(kind='bar')
plt.title('Avg Exam Score by Parental Education Level')
plt.show()
```

Parental Education Level also doesn't show much of a difference in exam scores across all three categories (bachelor, high school, masters). Parental Education Level does not impact on exam performance to that level.

```python
df1.groupby('internet_quality')['exam_score'].mean().plot(kind='bar')
plt.title('Avg Exam Score by Internet Quality')
plt.show()
```

Internet quality also shows no meaningful difference in exam scores all three categories average around 69-71.

```python
df1.groupby('extracurricular_participation')['exam_score'].mean().plot(kind='bar')
plt.title('Avg Exam Score by Extracurricular Participation')
plt.show()
```


Extracurricular participation shows absolutely no difference in exam scores both Yes and No average exactly 70.

Till now we can see that the feature which strongly impacts exam scores is "Study Hours", and moderate impact features are "Mental health rating" and "Exercise frequency" rest all have weak impact or non meaningful impact.

```python
df1['parental_education_level'].fillna(df1['parental_education_level'].mode()[0], inplace=True)
```

```python
print(df1.isnull().sum())
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
```

```python
cat_cols=['gender','part_time_job','diet_quality','parental_education_level','internet_quality','extracurricular_participation']
for col in cat_cols:
  df1[col]=le.fit_transform(df1[col])
```

```python
print(df1.head())
```

```python
df1=df1.drop(columns=['student_id'])
print(df1.shape)
```

```python
df2['Target']=df2['Target'].map({'Graduate':0,'Dropout':1})
print(df2['Target'].value_counts())
```

```python
X1=df1.drop(columns=['exam_score'])
y1=df1['exam_score']
```

```python
X2=df2.drop(columns=['Target'])
y2=df2['Target']
```

```python
print(X1.shape, y1.shape)
print(X2.shape, y2.shape)
```

```python
from sklearn.model_selection import train_test_split
X1_train, X1_test, y1_train, y1_test=train_test_split(X1, y1, test_size=0.2, random_state=42)
X2_train, X2_test, y2_train, y2_test=train_test_split(X2, y2, test_size=0.2, random_state=42)
print(X1_train.shape, X1_test.shape)
print(X2_train.shape, X2_test.shape)
```

for df1,

LINEAR REGRESSION

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, r2_score
```

```python
lr=LinearRegression()
lr.fit(X1_train, y1_train)
y1_pred_lr=lr.predict(X1_test)
```

```python
print("MAE: ",round(mean_absolute_error(y1_test, y1_pred_lr), 2))
```

```python
print('R2: ', round(r2_score(y1_test, y1_pred_lr), 2))
```

DECISION TREE

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
dt=DecisionTreeRegressor(random_state=42)
dt.fit(X1_train, y1_train)
y1_pred_dt=dt.predict(X1_test)
```

```python
print("MAE: ",round(mean_absolute_error(y1_test, y1_pred_dt), 2))
```

```python
print('R2: ', round(r2_score(y1_test, y1_pred_dt)))
```

After applying decision tree, we can see that it overfitted the data, hence it won't work for unseen data efficiently.

RANDOM FOREST

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
rf=RandomForestRegressor(n_estimators=100, random_state=42)
rf.fit(X1_train, y1_train)
y1_pred_rf=rf.predict(X1_test)
```

```python
print("MAE: ",round(mean_absolute_error(y1_test, y1_pred_rf), 2))
```

```python
print('R2: ', round(r2_score(y1_test, y1_pred_rf), 2))
```
